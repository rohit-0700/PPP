Program to display RF bands using if–elif

```python
f = float(input("Enter frequency in Hz: "))

if 30000 <= f <= 300000:
    print("VLF Band")
elif 300000 < f <= 3000000:
    print("LF Band")
elif 3000000 < f <= 30000000:
    print("MF Band")
elif 30000000 < f <= 300000000:
    print("HF Band")
else:
    print("Out of range")
```

Write a Python program that checks if a resistor value is within a tolerance range (e.g., 5% tolerance for a 100Ω resistor) and prints whether it is within tolerance or not.

```python
nominal_resistance = 100
tolerance_percent = 5
measured_resistance = 104

tolerance = nominal_resistance * (tolerance_percent / 100)
lower_limit = nominal_resistance - tolerance
upper_limit = nominal_resistance + tolerance

if lower_limit <= measured_resistance <= upper_limit:
    print("The resistor is within tolerance.")
else:
    print("The resistor is out of tolerance.")
```

```python
Q1: Largest of two numbers
IF a > b THEN
    PRINT "A is largest"
ELSE
    IF b > a THEN
        PRINT "B is largest"
    ELSE
        PRINT "Both equal"
    ENDIF
ENDIF
```

```python
a = int(input("Enter a: "))
b = int(input("Enter b: "))

if a > b:
    print("A is largest")
else:
    if b > a:
        print("B is largest")
    else:
        print("Both equal")
```

```python
Q2: Result with distinction
IF marks >= 40 THEN
    IF marks >= 75 THEN
        PRINT "Distinction"
    ELSE
        PRINT "Pass"
    ENDIF
ELSE
    PRINT "Fail"
ENDIF
```

```python
marks = int(input("Enter marks: "))

if marks >= 40:
    if marks >= 75:
        print("Distinction")
    else:
        print("Pass")
else:
    print("Fail")
```

```python
Q3: Login and role check
IF login = TRUE THEN
    IF role = "Admin" THEN
        PRINT "Admin Access"
    ELSE
        PRINT "User Access"
    ENDIF
ELSE
    PRINT "Login Failed"
ENDIF
```

```python
login = input("Login status (true/false): ")
role = input("Enter role: ")

if login == "true":
    if role == "Admin":
        print("Admin Access")
    else:
        print("User Access")
else:
    print("Login Failed")
```
