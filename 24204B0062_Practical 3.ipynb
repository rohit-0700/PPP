5. Sums of digits of a number

```python
num = int(input("Enter a number: "))
sum = 0 

while num > 0:
    sum += num % 10
    num //= 10

print("Sum of digits =", sum)
```

6. Reverse a number

```python
num = int(input("Enter a number: "))
rev = 0 

while num > 0:
    rev = rev * 10 + num % 10
    num //= 10
    print(rev)
```

Program to check if a number is a palindrome using for loop

```python
num = input("Enter a number: ")
rev =  ""

for i in num:
    rev = i + rev

if num == rev:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

Keep taking input until the user enters 0

Menu-driven calculator

```python
while True:
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Exit")

    ch = int(input("Enter choice: "))

    if ch == 4:
        break

    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))

    if ch == 1:
        print("Sum =", a + b)
    elif ch == 2:
        print("Difference =", a - b)
    elif ch == 3:
        print("Product =", a * b)
    else:
        print("Invalid choice")
```
